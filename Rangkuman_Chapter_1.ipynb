#Chapter 1: The Machine Learning Landscape
##1. Definisi Machine Learning
Machine Learning adalah ilmu (dan seni) pemrograman komputer sehingga mereka dapat belajar dari data tanpa harus diprogram secara eksplisit. ML adalah subset dari Artificial Intelligence (AI) yang memungkinkan sistem untuk secara otomatis belajar dan meningkatkan performa dari pengalaman.

##2. Mengapa Menggunakan Machine Learning?
###2.1 Masalah Tradisional Programming:
- Aturan kompleks dan sulit dipelihara
- Sulit beradaptasi dengan perubahan
- Membutuhkan banyak fine-tuning manual

###2.2 Keuntungan Machine Learning:
- Otomatis menemukan pola dalam data
- Adaptif terhadap perubahan data
- Dapat menangani masalah kompleks
- Memberikan insights tentang data

##3. Jenis-Jenis Machine Learning
###3.1 Supervised Learning
- Definisi: Algoritma belajar dari data yang sudah diberi label
- Contoh: Klasifikasi email spam, prediksi harga rumah
- Algoritma: Linear Regression, Decision Trees, Random Forest, SVM, Neural Networks

###3.2 Unsupervised Learning
- Definisi: Algoritma mencari pola dalam data tanpa label
- Contoh: Clustering customers, dimensionality reduction
- Algoritma: K-Means, Hierarchical Clustering, PCA, Autoencoders

###3.3 Reinforcement Learning
- Definisi: Agent belajar melalui trial and error dengan reward system
- Contoh: Game playing, robotics, autonomous vehicles
- Komponen: Agent, Environment, Actions, Rewards

##4. Batch vs Online Learning
###4.1 Batch Learning
- Melatih model dengan seluruh dataset sekaligus
- Tidak dapat belajar secara incremental
- Membutuhkan banyak resource untuk retraining

###4.2 Online Learning
- Belajar secara incremental dari data stream
- Adaptif terhadap perubahan data
- Efisien untuk big data

##5. Instance-based Vs Model-based Learning
###5.1 Instance-based Learning
- Mengingat contoh training data
- Prediksi berdasarkan similarity measure
- Contoh: k-Nearest Neighbors

###5.2 Model-based Learning
- Membangun model dari training data
- Prediksi menggunakan model yang sudah dilatih
- Contoh: Linear Regression, Neural Networks

##6. Tantangan Utama dalam Machine Learning
###6.1 Data Quality Issues
- Insufficient Training Data: Butuh data yang cukup untuk generalisasi yang baik
- Nonrepresentative Training Data: Data harus representatif terhadap populasi
- Poor-Quality Data: Missing values, outliers, noise dalam data
- Irrelevant Features: Feature engineering sangat penting

###6.2 Overfitting dan Underfitting
- Overfitting: Model terlalu kompleks, perform baik di training tapi buruk di test
- Underfitting: Model terlalu sederhana, perform buruk di training dan test
- Solutions: Regularization, cross-validation, more data

##7. Testing dan Validasi
###7.1 Train/Validation/Test Split
- Training Set: Data untuk melatih model
- Validation Set: Data untuk hyperparameter tuning
- Test Set: Data untuk evaluasi final

###7.2 Cross-Validation
- Teknik untuk evaluasi model yang lebih robust
- K-fold cross-validation adalah yang paling umum

##8. No Free Lunch Theorem
Tidak ada algoritma yang selalu terbaik untuk semua masalah. Pemilihan algoritma bergantung pada:
- Jenis data
- Ukuran dataset
- Kompleksitas masalah
- Interpretability requirements

```python
# Import libraries yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.datasets import make_classification, make_regression
from sklearn.metrics import accuracy_score, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

print("=== HANDS-ON MACHINE LEARNING CHAPTER 1 ===")
print("Demonstrasi berbagai konsep ML dasar\n")
```

```python
# 1. SUPERVISED LEARNING EXAMPLE - REGRESSION
print("1. SUPERVISED LEARNING - REGRESSION")
print("-" * 40)

# Generate sample regression data
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Split data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

# Train linear regression model
lr_model = LinearRegression()
lr_model.fit(X_train_reg, y_train_reg)

# Make predictions
y_pred_reg = lr_model.predict(X_test_reg)

# Evaluate
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f"Linear Regression MSE: {mse:.2f}")

# Visualize
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(X_train_reg, y_train_reg, alpha=0.6, label='Training Data')
plt.scatter(X_test_reg, y_test_reg, alpha=0.6, color='red', label='Test Data')
plt.plot(X_test_reg, y_pred_reg, color='green', linewidth=2, label='Predictions')
plt.title('Supervised Learning - Regression')
plt.legend()
```

```python
# 2. SUPERVISED LEARNING EXAMPLE - CLASSIFICATION
print("\n2. SUPERVISED LEARNING - CLASSIFICATION")
print("-" * 40)

# Generate sample classification data
X_clf, y_clf = make_classification(n_samples=100, n_features=2, n_redundant=0,
                                  n_informative=2, n_clusters_per_class=1, random_state=42)

# Split data
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

# Train KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train_clf, y_train_clf)

# Make predictions
y_pred_clf = knn_model.predict(X_test_clf)

# Evaluate
accuracy = accuracy_score(y_test_clf, y_pred_clf)
print(f"KNN Classification Accuracy: {accuracy:.2f}")

# Visualize
plt.subplot(1, 3, 2)
colors = ['red', 'blue']
for i in range(2):
    mask = y_train_clf == i
    plt.scatter(X_train_clf[mask, 0], X_train_clf[mask, 1],
               c=colors[i], alpha=0.6, label=f'Class {i}')
plt.title('Supervised Learning - Classification')
plt.legend()
```

```python
# 3. UNSUPERVISED LEARNING EXAMPLE - CLUSTERING
print("\n3. UNSUPERVISED LEARNING - CLUSTERING")
print("-" * 40)

# Use the same classification data but ignore labels
kmeans_model = KMeans(n_clusters=2, random_state=42, n_init=10)
clusters = kmeans_model.fit_predict(X_clf)

print(f"K-Means found {len(np.unique(clusters))} clusters")

# Visualize
plt.subplot(1, 3, 3)
colors = ['red', 'blue']
for i in range(2):
    mask = clusters == i
    plt.scatter(X_clf[mask, 0], X_clf[mask, 1],
               c=colors[i], alpha=0.6, label=f'Cluster {i}')
plt.scatter(kmeans_model.cluster_centers_[:, 0],
           kmeans_model.cluster_centers_[:, 1],
           c='black', marker='x', s=200, label='Centroids')
plt.title('Unsupervised Learning - Clustering')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 4. DEMONSTRASI OVERFITTING vs UNDERFITTING
print("\n4. OVERFITTING vs UNDERFITTING")
print("-" * 40)

# Generate polynomial data
np.random.seed(42)
X_poly = np.linspace(0, 1, 50).reshape(-1, 1)
y_poly = 1.5 * X_poly.ravel() + 0.5 * X_poly.ravel()**2 + 0.3 * np.random.randn(50)

# Polynomial features
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

# Different complexity models
models = {
    'Underfitting (degree=1)': Pipeline([
        ('poly', PolynomialFeatures(degree=1)),
        ('lr', LinearRegression())
    ]),
    'Good fit (degree=2)': Pipeline([
        ('poly', PolynomialFeatures(degree=2)),
        ('lr', LinearRegression())
    ]),
    'Overfitting (degree=15)': Pipeline([
        ('poly', PolynomialFeatures(degree=15)),
        ('lr', LinearRegression())
    ])
}

plt.figure(figsize=(15, 4))
X_plot = np.linspace(0, 1, 100).reshape(-1, 1)

for i, (name, model) in enumerate(models.items()):
    plt.subplot(1, 3, i+1)
    model.fit(X_poly, y_poly)
    y_plot = model.predict(X_plot)

    plt.scatter(X_poly, y_poly, alpha=0.6)
    plt.plot(X_plot, y_plot, color='red', linewidth=2)
    plt.title(name)
    plt.xlabel('X')
    plt.ylabel('y')

plt.tight_layout()
plt.show()
```

```python
print("\n5. TRAIN/VALIDATION/TEST SPLIT DEMONSTRATION")
print("-" * 40)

# Demonstrate proper data splitting
from sklearn.model_selection import validation_curve

X_demo, y_demo = make_regression(n_samples=200, n_features=1, noise=10, random_state=42)

# Split into train+validation and test
X_temp, X_test_final, y_temp, y_test_final = train_test_split(
    X_demo, y_demo, test_size=0.2, random_state=42
)

# Split train+validation into train and validation
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=42  # 0.25 * 0.8 = 0.2 of total
)

print(f"Total samples: {len(X_demo)}")
print(f"Training samples: {len(X_train_final)} ({len(X_train_final)/len(X_demo)*100:.1f}%)")
print(f"Validation samples: {len(X_val)} ({len(X_val)/len(X_demo)*100:.1f}%)")
print(f"Test samples: {len(X_test_final)} ({len(X_test_final)/len(X_demo)*100:.1f}%)")
```

```python
# 6. CROSS-VALIDATION EXAMPLE
print("\n6. CROSS-VALIDATION EXAMPLE")
print("-" * 40)

from sklearn.model_selection import cross_val_score

# Compare different algorithms using cross-validation
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

models_cv = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=10, random_state=42),
    'SVR': SVR(kernel='rbf', C=1.0)
}

print("Cross-validation scores (R²):")
for name, model in models_cv.items():
    scores = cross_val_score(model, X_temp, y_temp, cv=5, scoring='r2')
    print(f"{name}: {scores.mean():.3f} (+/- {scores.std() * 2:.3f})")

print("\n=== CHAPTER 1 COMPLETE ===")
print("Key takeaways:")
print("1. ML learns patterns from data without explicit programming")
print("2. Three main types: Supervised, Unsupervised, Reinforcement")
print("3. Main challenges: Data quality, Overfitting/Underfitting")
print("4. Proper evaluation requires train/validation/test splits")
print("5. No single algorithm works best for all problems")
```

##**14. Dapatkah Anda menyebutkan empat tantangan utama dalam Machine Learning?**

**Jawban:**
- Data Training Tidak Cukup: Tidak cukup contoh untuk mempelajari pola secara efektif
- Kualitas Data Buruk: Nilai yang hilang, error, outlier, fitur yang tidak relevan
- Overfitting: Model bekerja baik pada data training tetapi buruk pada data baru
- Underfitting: Model terlalu sederhana untuk menangkap pola yang mendasari

##**15. Jika model Anda bekerja sangat baik pada data training tetapi generalisasi buruk pada instance baru, apa yang terjadi? Dapatkah Anda menyebutkan tiga solusi yang mungkin?**

**Jawban:**

Masalah: Overfitting - model telah menghafal data training daripada belajar pola yang dapat digeneralisasi.

Solusi:
- Regularisasi: Menambahkan istilah penalti untuk mencegah kompleksitas model
- Dapatkan lebih banyak data training: Contoh yang lebih beragam membantu generalisasi
- Kurangi kompleksitas model: Gunakan algoritma yang lebih sederhana atau lebih sedikit fitur

##**16. Apa itu test set, dan mengapa Anda ingin menggunakannya?**

**Jawban:**

Test set adalah bagian dari data yang ditahan selama training dan hanya digunakan untuk evaluasi model akhir. Ini memberikan estimasi yang tidak bias tentang performa model pada data yang belum pernah dilihat, mensimulasikan kondisi deployment dunia nyata. Ini penting untuk menilai kemampuan generalisasi dan menghindari overfitting pada data validasi.

##**17. Apa tujuan dari validation set?**

**Jawban:**

Validation set digunakan selama pengembangan model untuk:
- Hyperparameter tuning: Memilih pengaturan konfigurasi terbaik
- Model selection: Memilih antara algoritma yang berbeda
- Early stopping: Mencegah overfitting selama training
- Performance monitoring: Melacak generalisasi selama pengembangan

##**18. Apa itu train-dev set, kapan Anda membutuhkannya, dan bagaimana Anda menggunakannya?**

**Jawban:**

Train-dev set adalah bagian dari data training yang digunakan untuk mendeteksi apakah ada ketidakcocokan antara distribusi data training dan validasi/test.

Kapan dibutuhkan: Ketika data training dan test berasal dari sumber/distribusi yang berbeda

Cara menggunakan:
- Bagi data training menjadi train dan train-dev
- Jika model bekerja baik pada train-dev tetapi buruk pada validasi, ada masalah ketidakcocokan data
- Jika model bekerja buruk pada keduanya, itu masalah overfitting/underfitting

##**19. Apa yang bisa salah jika Anda menyetel hyperparameter menggunakan test set?**

**Jawban:**

Terjadi data leakage - Anda secara tidak langsung melatih pada test set, yang menyebabkan:
- Estimasi performa yang terlalu optimis
- Generalisasi buruk pada data yang benar-benar belum pernah dilihat
- Evaluasi model yang tidak valid
- Test set tidak lagi mewakili performa dunia nyata

Solusi: Gunakan validation set terpisah untuk hyperparameter tuning, jaga test set tetap tidak tersentuh sampai evaluasi akhir.
